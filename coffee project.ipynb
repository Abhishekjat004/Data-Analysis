```python
import pandas as pd
total_prod = pd.read_csv(r'D:\AU(DS)\python\coffe\total-production.csv')
```

```python
total_prod.info()
```

```python
total_prod.head()
```

```python
add_to_columns = total_prod.iloc[:,1:]
add_to_columns.head()
```

```python
total_prod['sum_Production'] = add_to_columns.sum(axis=1)
```

```python
total_prod.loc[:,['total_production','sum_Production']].round().head()
```

```python
top_10 = total_prod.loc[:,['total_production','sum_Production']].round().sort_values(ascending=False,by="sum_Production").iloc[:10]
```

```python
top_10
```

```python
top_10.set_index('total_production',inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
fig,ax = plt.subplots()
ax.bar(top_10.index,top_10.sum_Production);
ax.set_ylabel("Productions in (millions)")
plt.xticks(rotation= 45);
```

```python
top_10.index
```

```python
top_2012 = total_prod.sort_values(ascending=False,by="2012").iloc[:5]
top_2012 = top_2012.loc[:,['total_production','2012']]
```

```python
fig,ax = plt.subplots()
ax.bar(top_2012['total_production'],top_2012['2012']);
```

```python
total_prod.drop("sum_Production",axis=1,inplace=True)
```

```python
df_transposed = total_prod.transpose()
```

```python
df_transposed.head()
```

```python
df_transposed.columns=df_transposed.iloc[0]
```

```python
df_transposed.head()
```

```python
df_transposed = df_transposed.drop(df_transposed.index[0])           #remove first row
```

```python
df_transposed.head()
```

```python
fig,ax=plt.subplots()
ax.plot(df_transposed.index,df_transposed.loc[:,['Angola','Brazil','Colombia','Viet Nam']]);
ax.legend(['Angola','Brazil','Colombia','Viet Nam'])
plt.xticks(rotation=90);
ax.set_title("Coffe Production during 1990 - 2018");
```

```python
other_countries = df_transposed
other_countries = other_countries.assign(rest_of_world=other_countries.drop("Brazil",axis=1).sum(axis=1)).loc[:,['Brazil','rest_of_world']]
```

```python
other_countries.head()
```

```python
other_countries = other_countries.astype({'Brazil':float,'rest_of_world':float})
```

```python
other_countries.info()
```

```python
fig,ax = plt.subplots();
ax.stackplot(other_countries.index,other_countries.Brazil,other_countries.rest_of_world);
plt.xticks(rotation=90);
ax.set_title("Brazil vs Other Countries Production");
ax.legend(other_countries.columns,loc='upper left');
ax.spines['right'].set_visible(False);
ax.spines['top'].set_visible(False);
```

```python
fig,ax = plt.subplots()
ax.scatter(df_transposed.Brazil,df_transposed['Angola']);
ax.set_title("Brazil vs Angola Production");
```

```python
rest_of_world = df_transposed.copy()
rest_of_world = rest_of_world.assign(rest_of_world=rest_of_world.drop(['Brazil','Viet Nam','Colombia','Indonesia','Ethiopia'], axis = 1).sum(axis=1))
```

```python
rest_of_world = rest_of_world.loc[:,['Brazil',  'Viet Nam','Colombia',  'Indonesia', 'Ethiopia','rest_of_world']]
```

```python
fig,ax = plt.subplots()
ax.pie(x = rest_of_world.loc['2018',:],
      startangle=90,labels=rest_of_world.columns,autopct="%.0f%%");
ax.set_title("Comparing the countries in Production");
```
